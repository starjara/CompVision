# 1.5.3 Shi-Tomasi Corner Detector & Good Features to Track

<h3>
    <ul>
        <li>We will learn about the another corner detector : Shi-Tomasi Corner Detector</li>
        <li>We will see the functions : cv2.goodFeatureToTrack</li>
    </ul>
    </h3>


```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('horse.bmp')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

corners = cv2.goodFeaturesToTrack(gray, 25, 0.01, 10)
corners = np.int0(corners)

for i in corners :
    x, y = i.ravel()
    cv2.circle(img, (x, y), 3, 255, -1)
    
plt.figure(figsize=(20,20)) 
plt.imshow(img), plt.show()
```
