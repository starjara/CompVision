#### Scaling

```python
import cv2
import numpy as np

img = cv2.imread ('sIMG_2168.jpg')

res = cv2.resize (img, None, fx=8, fy=8, interpolation = cv2.INTER_NEAREST)
res_cubic = cv2.resize (img, None, fx=8, fy=8, interpolation = cv2.INTER_CUBIC)
# OR

height, width = img.shape [:2]
res2 = cv2.resize (img, (2*width, 2*height), interpolation = cv2.INTER_CUBIC)

while (1) :
    cv2.imshow ('img', img)
    cv2.imshow ('res', res)
    cv2.imshow ('res2', res2)
    cv2.imshow ('res_cubic', res_cubic)
    
    k =cv2.waitKey (1) & 0xFF
    if (k == 27) :
        break
        
cv2.destroyAllWindows()
```

#### Translation

```python
import cv2
import numpy as np

img = cv2.imread ('sIMG_2168.jpg')
rows, cols, ch = img.shape

tx, ty = 150, 300

M = np.float32 ([[1, 0, tx], [0, 1, ty]])
dst = cv2.warpAffine (img, M, (cols, rows))
dst = cv2.rectangle (dst, (0, 0), (tx, ty), (0, 255, 0), -1)

cv2.imshow ('img', dst)
cv2.waitKey (0)
cv2.destroyAllWindows ()
```

#### Rotation

```python
import cv2
import numpy as np

img = cv2.imread ('sIMG_2164.jpg', 1)
rows, cols, ch = img.shape

cx, cy = 100, 0
theta = 10

M = cv2.getRotationMatrix2D ((cx, cy), theta, 1)
dst = cv2.warpAffine (img, M, (cols, rows))

dst = cv2.circle (dst, (cx, cy), 50, (0, 255, 0), -1)

cv2.imshow ('img', dst)
cv2.waitKey (0)
cv2.destroyAllWindows ()
```

#### Affine Transformaion

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread ('dsu4.jpg')
rows, cols, ch = img.shape

pts1 = np.float32 ([[235, 457], [238, 269], [569, 231]])
pts2 = np.float32 ([[235, 457], [235, 270], [570, 270]])

M = cv2.getAffineTransform (pts1, pts2)

dst = cv2.warpAffine (img, M, (cols, rows))

plt.subplot (121), plt.imshow (img), plt.title('Input')
plt.subplot (122), plt.imshow (dst), plt.title('Output')
plt.show()
```

#### Perspective Tranformation

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread ('dsu4.jpg')
rows, cols, ch = img.shape

pts1 = np.float32 ([[235, 457], [238, 269], [569, 231], [564, 493]])
pts2 = np.float32 ([[235, 457], [235, 270], [570, 270], [570, 457]])

M = cv2.getPerspectiveTransform (pts1, pts2)
dst = cv2.warpPerspective (img, M, (cols, rows))

plt.figure (figsize = (10, 15))

plt.subplot (121), plt.imshow (img), plt.title ('Input')
plt.subplot (122), plt.imshow (dst), plt.title('Output')
plt.show()

door = cv2.imread ('dsu5.jpg')
rows, cols, ch = door.shape

pts1 = np.float32 ([[578, 336], [756, 378], [586, 544], [767, 544]])
pts2 = np.float32 ([[578, 336], [756, 336], [576, 544], [757, 544]])

M = cv2.getPerspectiveTransform (pts1, pts2)
dst = cv2.warpPerspective (door, M, (cols, rows))

plt.figure (figsize = (10, 15))

plt.subplot (121), plt.imshow (door), plt.title ('Input')
plt.subplot (122), plt.imshow (dst), plt.title('Output')
plt.show()

paper = cv2.imread ('paper.png')
rows, cols, ch = paper.shape

pts1 = np.float32 ([[285, 476], [1038, 492], [1010, 1597], [215, 1548]])
pts2 = np.float32 ([[285, 476], [1038, 476], [1038, 1597], [285, 1597]])

M = cv2.getPerspectiveTransform (pts1, pts2)
dst = cv2.warpPerspective (paper, M, (cols, rows))

plt.figure (figsize = (10, 15))

plt.subplot (121), plt.imshow (paper), plt.title ('Input')
plt.subplot (122), plt.imshow (dst), plt.title('Output')
plt.show()
```
