## Code
<h3>
    In all the above functions, you will see some common arguments as given below :
    <ul>
        <li>img : The image where you want to draw the shapse</li>
        <li>color : Color of the shape. for BGR, pas it as a tuple, eg : (255, 0, 0) for blue. For grayscale, just pass the scalar value.</li>
        <li>thickness : Thickness of the line or circle etc. If -1 is passed for colsed figures like circles, it will fill the shape <i>defulat thickness = 1</i></li>
        <li>lineType : Type of line, whether 8-connected, anti-aliased line etc. By default, it is 8-connected. <code>cv2.LINE_AA</code> gives anti-aliased line which looks great for curves
    </ul>
</h3>

## Drawing Line
<h3> To draw a line, you need to pass starting and ending coordinates of line. We will create a black image and draw a blue line on it from top-left to bottom-right corners.

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

# Create a black image
img = np.zeros ((512, 512, 3), np.uint8)

# Draw a diagonal blue line with thickness of 5 px
img = cv2.line (img, (0,0), (511, 511), (255, 0, 0), 5)
plt.imshow (img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])    # to hide tick values on X
plt.show()
```

## Draw Rectangle

<h3>
    To draw a rectangle, you neep top-left corner and bottom-right corner of rectangle. This time we will draw a green rectangle at the top-right corner of image.    
</h3>

```python
img = cv2.rectangle (img, (384, 0), (510, 218), (0, 255, 0), 3)
plt.imshow (img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])    # to hide tick values on X
plt.show()
```

## Drawing Circle
<h3>
    To draw a circle, you need its center coordinates and radius. We will draw a circle inside the rectangle drawn above.
</h3>

```python
img = cv2.circle (img, (447, 63), 63, (0,0,255), -1)
plt.imshow (img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])    # to hide tick values on X
plt.show()
```

## Drwing Ellipse
<h3>
    To draw the ellipse, we need to pass several arguments. One argument is the center location (x, y). Next argument is axes lenghts (major axis length, minor axis length). angle is the angle of rotation of ellipse in anti-clockwise directinon. startAngle and endAngle denotes the starting and ending of ellipse arc measured in clockwise direction from major axis. i.e. giving values 0 and 360 gives the full ellipse. For more details, check the documentation of cv2.ellipse(). Below example draws a half ellips at the center of the image.
</h3>

```python
img = cv2.ellipse (img, (256, 256), (100, 50), 0, 0, 270, 255, -1)
plt.imshow (img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])    # to hide tick values on X
plt.show()
```

## Drawing Polygon
<h3>
    To draw a polygon, first you need coordinates of vertices. Make those points into an array of shape ROWSx1x2 where ROWS are number of vertices and it should be of type int32. Here we draw a small polygon of with four vertices in yellow color.
</h3>

```python
pts = np.array ([[10,5], [20,30], [70, 20], [50, 10]], np.int32)
pts = pts.reshape ((-1,1,2))
img = cv2.polylines (img, [pts], True, (0,255,255))
plt.imshow (img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])    # to hide tick values on X
plt.show()
```

### Note : If thirt argument is False, you will get a polylines joining all the points, not a colsed shape.
### Note : cv2.polylines () can be used to draw multiple lines. Just create a list of all the lines you want to draw and pass it to the function. All lines will be drawn individually. It is more better and faster way to draw a group of lines than calling cv2.lines() for each line.

## Adding Text to Images :
<h3>
    To put thexts in images, you need specify following things.
    <ul> 
        <li>Text data that you wnat to write</li>
        <li>Position coordinates of where you want put it (i.e. bottom-left corner where data starts)</li>
        <li>Font type (check cv2.putText() docs for supported fonts_</li>
        <li>Fonst Scale (specifies the size of font)</li>
        <li>regular things like color, thickness, lineType etc. For better look, lineType = cv2.LINE_AA is recommended</li>
    </ul>
    We will write OpenCV on our image in white color.
</h3>

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText (img, 'OpenCv', (10, 500), font, 4, (255, 255, 255), 2, cv2.LINE_AA)
plt.imshow (img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])    # to hide tick values on X
plt.show()
```
