# 1.4.16 Interactive Foreground Extraction using GrabCut Algorithm

<h3>
    Goals
    <ul>
        <li>We will see GrabCut algorithm to extract foreground in image</li>
        <li>We will create an interactive application for this</li>
    </ul>
</h3>

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('horse.bmp')
mask = np.zeros(img.shape[:2], np.uint8)

bgdModel = np.zeros((1, 65), np.float64)
fgdModel = np.zeros((1, 65), np.float64)

rect = (378, 88, 780, 570)
cv2.grabCut(img, mask, rect, bgdModel, fgdModel, 5, cv2.GC_INIT_WITH_RECT)

mask2 = np.where((mask == 2) | (mask == 0), 0, 1).astype('uint8')
img = img*mask2[:,:,np.newaxis]

plt.figure(figsize=(12, 12))
plt.imshow(img), plt.colorbar(), plt.show()
```
